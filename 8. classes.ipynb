# Classes

A **class** is a template or blueprint: essentially, it is quite similar to defining a 'type' of which one or more objects can be created. 

It is also known as an **Abstraction** or an Abstract Datatype.

It **encaspsulates** data as well as methods.

A class is declared as follows:

class class_name:

    Functions
    Variables

## UML (Unified Modelling Language) Class Digram


```python
class ProductClass:
    pass
```

**pass** in python means do nothing. 

Above, a class object named "ProductClass" is declared. 

Now consider an "egclass" which has all the characteristics of "ProductClass". So all you have to do is, equate the "egclass" to "ProductClass". 
In python jargon this is called as creating an instance. "egclass" is the **instance** of "ProductClass". Creating an object from class is also called **instantiation**

Once instantiated, an object has a unique identity (address in memory), a state which may change(the values of variables it stores) and an associated behavior (methods defined within).

This concept is known as Object Oriented Programming.

```python
egclass = ProductClass()
```

```python
type(egclass)
```

```python
type(ProductClass)
```

Now let us add some "functionality" to the class. So that our "FirstClass" is defined in a better way. A function inside a class is called as a **"Method"** of that class

The first parameter **self** is a reference to the object itself. We can give this any other name, but the convention is to use *self*. This is compulsory for all methods.
An object's self parameter is not visible to other objects

```python
class ProductClass:
    def fn(self,name,price):
        print(name,price)
```

```python
eg0 = ProductClass()
eg0.fn("Book","500")
```

Classes will have a function named **"\_\_init\_\_"**. These are called as **magic methods**. In this method you basically initialize the variables of that class or any other initial functionality.

It is also known as class **constructor** and is **automatically** called whenever an object is instantiated.

Without the use of magic method or \_\_init\_\_ which is otherwise called as **constructor**. One had to define an **init( )** method and call the **init( )** function.

```python
class Product:
    def __init__(self,name,price):
        self.name = name
        self.price = price
```

```python
eg1 = Product('book',500)
eg2 = Product('pen',20)
```

```python
print(eg1.name, eg1.price)
print(eg2.name, eg2.price)
```

self.name and self.price create two variables: name and price which are known as class **attributes**

Attributes are variables that store data for the objects.
